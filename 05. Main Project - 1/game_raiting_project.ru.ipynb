**3. Столбец `year_of_release`**

Заметим, что некоторые названия игр имеют год. Присвоим этот год году выпуска игры.

```python
df['year_of_release'].isna().sum()
```

```python
def name_year_parcer(row):
    if pd.isna(row['year_of_release']):
        for i in row['name'].split():
            if i.isdigit() and int(i) >= 1980:
                return int(i)
    else:
        return row['year_of_release']
        
df['year_of_release'] = df.apply(name_year_parcer, axis=1)
```

```python
print(
    df['year_of_release'].isna().sum(),
    df['year_of_release'].isna().sum() / len(df) * 100
)
```

Восстановить остальные даты выпуска игр по имеющимся данным невозможно. Пропущенных данных 254 строки. Это 1,5% от всей информации, поэтому удалим пропущенные значения.

# Проверка гипотез

## Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

[*Вернуться к оглавлению*](#Оглавление)

Нулевая гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC равны.  
Альтернативная гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC не равны.

```python
# критический уровень статистической значимости
alpha = 0.05

# Т-тест о равенстве средних двух выборок
# платформа Xbox One была выпущена в 2013 году, поэтому сравниваем выборки за одинаковый период
results = st.ttest_ind(
    df[df['platform']=='XOne']['user_score'].dropna(),
    df[(df['platform']=='PC') & (df['year_of_release']>=2013)]['user_score'].dropna(),
    equal_var = False 
)

print(
    'Средний рейтинг платформы Xbox One:', 
    df[df['platform']=='XOne']['user_score'].mean()
)
print(
    'Средний рейтинг платформы PC:', 
    df[df['platform']=='PC']['user_score'].mean()
)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Полученное Р-значение больше уровня статистической значимости. Т.е. вероятность получить такое или большее различие между средними случайно почти 15% и гипотеза о равенстве между средними пользовательскими рейтингами платформ Xbox One и PC не может быть отвергнута.

## Средние пользовательские рейтинги жанров Action и Sports разные

[*Вернуться к оглавлению*](#Оглавление)

Нулевая гипотеза: Средние пользовательские рейтинги жанров Action и Sports равны.  
Альтернативная гипотеза: Средние пользовательские рейтинги жанров Action и Sports не равны.

```python
# критический уровень статистической значимости
alpha = 0.05

# Т-тест о равенстве средних двух выборок
results = st.ttest_ind(
    df[(df['genre']=='Action') & (df['year_of_release']>=2013)]['user_score'].dropna(),
    df[(df['genre']=='Sports') & (df['year_of_release']>=2013)]['user_score'].dropna(),
    equal_var = False 
)

print(
    'Средний рейтинг жанра Action:', 
    df[df['genre']=='Action']['user_score'].mean()
)
print(
    'Средний рейтинг жанра Sports:', 
    df[df['genre']=='Sports']['user_score'].mean()
)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Полученное значение p-value меньше уровня статистической значимости. Т.е. маловероятна случайность полученных результатов, что дает основание отвергнуть гипотезу, что средние рейтинги пользователей жанров Action и Sports равны.
