1. Исследуем столбец с наименованиями населенных пунктов `locality_name`.

```python
df['locality_name'].unique()
```

В списке присутствуют дубликаты названий населенных пунктов. Например: 'посёлок Мурино', 'поселок Мурино' и 'Мурино'. Часто встречаются дубликаты `посёлок` и `поселок`.

Во всех названиях населенных пунктов проведем замену `поселок` на `посёлок`.
Также создадим отдельный столбец `city_name`, который будет содержать только названия населенных пунктов.

```python
# Заменим NaN на None
df['locality_name'] = df['locality_name'].fillna(value='None')

# Заменим посел на посёл
df['locality_name'] = df['locality_name'].str.replace('посел', 'посёл')

# Функция, которая отделяет тип населенного пункта от названия 
def city_name(val):
    if ('посёлок городского типа' in val):
        return val.replace('посёлок городского типа ', '')
    if ('городской посёлок' in val):
        return val.replace('городской посёлок ', '')
    if ('коттеджный посёлок' in val):
        return val.replace('коттеджный посёлок ', '')    
    if ('посёлок' in val):
        return val.replace('посёлок ', '')  
    if ('деревня' in val):
        return val.replace('деревня ', '')  
    if ('садовое товарищество' in val):
        return val.replace('садовое товарищество ', '')          
    if ('село' in val):
        return val.replace('село ', '')  
    if ('садоводческое некоммерческое товарищество' in val):
        return val.replace('садоводческое некоммерческое товарищество ', '')    
    return val

# Создадим столбец city_name с названием городов
df['city_name'] = df['locality_name'].apply(city_name)
df['city_name'].value_counts().head(10)
```

```python
print('Количество неизвестных населенных пунктов', len(df[df['locality_name']=='None']))
```

Восстановить названия населенных пунктов по имеющимся данным невозможно. Причину отсутствия данных установить невозможно. Необходимо сообщить ответственному отделу об отсутствии информации по населенным пунктам.

3. Заполним пропуски в столбце `living_area` - жилая площадь. Пропуски составляют 8% данных.  

Проверим корреляцию между общей площадью квартиры и жилой площадью.

```python
# Рассчитаем коэффицент корреляции
print(df['living_area'].corr(df['total_area']))

# Построим диаграмму рассеяния
df.plot(x='living_area', y='total_area', kind='scatter');
```

Коэффициент корреляции высокий, равен 0,94. Значит есть прямая связь между общей и жилой площадью.

Чтобы заполнить пропуски, разобьем общую площадь на группы: 0-25 кв.м, 25-50 кв.м, 50-75 кв.м, 75-100 кв.м, 100-150 кв., 150-200 кв.м. Для каждой группы вычислим отношение медиан общей и жилой площади. Для каждого пропуска в жилой площади разделим общую площадь на найденный коэффициент `area_ratio`.

```python
# Разобъем общую площадь по группам
cut_labels = ['25', '50', '75', '100', '150', '200']
cut_bins = [0, 25, 50, 75, 100, 150, 1000]
df['area_group'] = pd.cut(df['total_area'],
                             bins=cut_bins,
                             labels=cut_labels)

# Определим отношение медиан общей площади к жилой для каждой группы
area_ratio = df.groupby('area_group')['total_area'].median() / df.groupby('area_group')['living_area'].median()

# Создадим датафрейм с коэффициентом площадей
area_ratio = pd.DataFrame(area_ratio)

# Переименуем колонку в датафрейме
area_ratio.columns = ['area_ratio']

# Объединим общий датафрейм с датафреймом area_ratio
df = df.join(area_ratio, on='area_group')

# Рассчитаем пропуски в жилой площади через коэффицент площадей
df['living_area'] = df['living_area'].fillna(df['total_area'] / df['area_ratio'])
```

Были заполнены пропуски в столбце с данными о жилой площади. Пропуски могли проявиться из-за того, что пользователи не внесли данную информацию в объявление. Однако по имеющимся данным о общей плоащди жилья удалось восстановить пропущенные значения.

4. Заполним пропуски в столбце `kitchen_area`. Пропуски составляют 9.6% данных.

Проверим корреляцию между общей площадью квартиры и жилой площадью.

```python
# Рассчитаем коэффицент корреляции
print(df['kitchen_area'].corr(df['total_area']))

# Построим диаграмму рассеяния
df.plot(x='kitchen_area', y='total_area', kind='scatter');
```

Корреляция не очень. Построим зависимость медианных значений площади кухни и общей площади.

```python
df.groupby('area_group')[['total_area', 'kitchen_area']].median().plot(x='kitchen_area', y='total_area');
```

На построенном графике лучше прослеживается зависимость между общей площадью и площадью кухни.  
Чтобы заполнить пропуски, воспользуемся ранее проделанной разбивкой на группы общей площади. Для каждой группы вычислим отношение медиан общей площадью и кухней. Для каждого пропуска в площади кухни разделим общую площадь на найденный коэффициент `kitchen_ratio`.

```python
# Воспользуемся группами area_group найденными раньше
# Определим отношение медиан общей площади к площади кухни для каждой группы
kitchen_ratio = (
    df.groupby('area_group')['total_area'].median() / 
    df.groupby('area_group')['kitchen_area'].median()
)

# Создадим датафрейм с коэффициентом площадей
kitchen_ratio = pd.DataFrame(kitchen_ratio)

# Переименуем колонку в датафрейме
kitchen_ratio.columns = ['kitchen_ratio']

# Объединим общий датафрейм с датафреймом kitchen_ratio
df = df.join(kitchen_ratio, on='area_group')

# Заполним пропуски в площади кухни через коэффицент площадей
df['kitchen_area'] = df['kitchen_area'].fillna(df['total_area'] / df['kitchen_ratio'])
```

Были заполнены пропуски в столбце с данными о площади кухни. Пропуски, возможно, были допущены из-за того, что пользователи не указали значения при оформлении объявления. Однако имеющиеся данные по общей площади квартир, позволили заполнить пропуски.

5. Столбец с количеством дней до продажи `days_exposition` имеет пропущенные значения. Количесвто пропусков 13,4%. Скорей всего это означает, что объявление о продаже до сих пор действительно.  
Заменим пропуски на разницу между датой загрузки объявления и датой выгрузки базы. <a id='date'></a>

```python
# Изменим формат столбца с датой на datetime
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%d')

# Вычислим количество дней от момента публикации до выгрузки базы
df['delta_day'] = df['first_day_exposition'].max() - df['first_day_exposition']

# Заполним пропуски в days_exposition из столбца delta_day
df['days_exposition'] = df['days_exposition'].fillna(df['delta_day'].dt.days)
```

В данных есть аномальные величины: высота потолка не может быть 1м и 100м. При этом средние и медианные значения 2,8м и 2,6м соответсвенно выглядят правдоподобно.  
Исследуем строки с аномально большими значениями высоты потолка. Аномальными будем считать потолки ниже двух метров и выше шести метров, предполагая, что есть, например, двухэтажные квартиры.

```python
# Выделим строки с высотой более 6м
df.query('ceiling_height > 6')
```

```python
# Выделим строки с высотой ниже 2м
df.query('ceiling_height < 2.0')
```

Исправим данные в столбце `ceiling_heiqht`:  
1. Значения больше 20 и меньше 40 разделим на 10, т.к. вероятно, что ошиблись знаком при внесении данных.
2. Значения больше 6 и меньше 19, а также 100, ошибочные. Заменим их на NaN.
3. Значения меньше двух тоже принимаем как ошибочные. Заменим их на NaN.
4. Заменим NaN на медианные значения.

```python
# Заменим пропуски в ceiling_heiqht в диапазоне между 20 и 40
df['ceiling_height'] = df['ceiling_height'].where(
    (df['ceiling_height'] < 20) | (df['ceiling_height'] > 40), df['ceiling_height'] / 10
)

# Заменим пропуски в ceiling_heiqht в диапазоне между 6 и 19
df['ceiling_height'] = df['ceiling_height'].where(
    (df['ceiling_height'] < 6) | (df['ceiling_height'] > 19), np.nan
)

# Заменим пропуски в ceiling_heiqht < 2
df['ceiling_height'] = df['ceiling_height'].where(
    (df['ceiling_height'] >= 2), np.nan
)

# Заменим в ceiling_heiqht значение 100 на NaN
df['ceiling_height'] = df['ceiling_height'].replace(100, np.nan)

# Заменим оставшиеся пропуски на медианные значения
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
```

### День недели, месяц и год публикации объявления

[*Вернуться к оглавлению*](#Оглавление)

```python
# День недели публикации объявления
df['day_exposition'] = df['first_day_exposition'].dt.day_name()

# Месяц подачи объявления
df['month_exposition'] = df['first_day_exposition'].dt.month_name()

# Год подачи объявления
df['year_exposition'] = df['first_day_exposition'].dt.year
```

### Этаж квартиры: первый, последний, другой.

[*Вернуться к оглавлению*](#Оглавление)

```python
# Функция определяющая этаж: первый, последний, другой.
def floor_type(row):
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] - row['floors_total'] == 0:
        return 'последний'
    return 'другой'

df['floor_type'] = df.apply(floor_type, axis=1)
```

[*Вернуться к оглавлению*](#Оглавление)

```python
# соотношение жилой и общей площади
df['area_ratio'] = df['living_area'] / df['total_area']

# отношение площади кухни к общей
df['kitchen_ratio'] = df['kitchen_area'] / df['total_area']
```
